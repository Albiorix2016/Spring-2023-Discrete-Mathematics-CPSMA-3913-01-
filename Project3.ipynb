We were asked to create our opwn sample space, , compute probabilities and simulate selecting items from it.

The First part asked us to create a sample space with 3 or more simple events. I decided to create 3 events from the Harry Potter universe. I will assign a probablity to each space then show the axioms of probability hold.

```python
HarryPotter_SampleSpace = {
  'Quidditch Game Played': 0.3, 
  'Bufferfly Collected': 0.1,
  'House Cup Won': 0.2,
  'Marauder's Map Found': 0.4
}

# The axioms of probability hold: 
# 1. All probabilities are between 0 and 1: 
for key in HarryPotter_SampleSpace:
    assert HarryPotter_SampleSpace[key] >= 0 and HarryPotter_SampleSpace[key] <= 1

# 2. The sum of all the probabilities is 1
total = 0
for key in HarryPotter_SampleSpace:
    total += HarryPotter_SampleSpace[key]
assert total == 1.0
```

We were then asked to create a compound event that uses some of the events that we came up with

```python
Compound_Event = {
    'Quidditch Game Played': HarryPotter_SampleSpace['Quidditch Game Played'], 
    'Marauder's Map Found': HarryPotter_SampleSpace['Marauder's Map Found']
}
```

We were then tasked with computing the porbablity of a union and intersection of the events

```python
# Probability of union of Quidditch Game Played & Marauder's Map Found
Probability_union = HarryPotter_SampleSpace['Quidditch Game Played'] + HarryPotter_SampleSpace['Marauder's Map Found'] - (
    HarryPotter_SampleSpace['Quidditch Game Played'] * HarryPotter_SampleSpace['Marauder's Map Found'])

# Probability of intersection of Quidditch Game Played & Marauder's Map Found
Probability_intersection = HarryPotter_SampleSpace['Quidditch Game Played'] * HarryPotter_SampleSpace['Marauder's Map Found']
```

Next we need to compute both the conditional probabilities using 2 events. Then apply  Bayes theorem to verify the results

```python
# P(A  B) / P(B)
Conditional_Probability_1 = HarryPotter_SampleSpace['Quidditch Game Played'] * HarryPotter_SampleSpace['Marauder's Map Found'] /  HarryPotter_SampleSpace['Marauder's Map Found']

# P(B  A) / P(A)

Conditional_Probability_2 = HarryPotter_SampleSpace['Quidditch Game Played'] * HarryPotter_SampleSpace['Marauder's Map Found'] /  HarryPotter_SampleSpace['Quidditch Game Played']

# Bayes theorem verification: 
assert Conditional_Probability_1 == Conditional_Probability_2

# the conditional probability between two events A and B is the probability of observing both events A and B given that event A has occurred. 
```

Next we need to compute the probability of the coplement of an event. I picked Marauder's Map Found.

```python
# Probability of complement of Marauder's Map Found
Probability_complement = 1 - HarryPotter_SampleSpace['Marauder's Map Found']
```

Finally we were tasked with simulating a simple occurrence of your sample space 10 times. Pick an event, does the simulation give you exactly the same probability as your assignment? Explain why or why not

```python
import random

simulated_results = []

for i in range(10):
    num = random.random()
    if num <= HarryPotter_SampleSpace['Quidditch Game Played']:
        simulated_results.append('Quidditch Game Played')
    elif num <= HarryPotter_SampleSpace['Quidditch Game Played'] + HarryPotter_SampleSpace['Bufferfly Collected']:
        simulated_results.append('Bufferfly Collected')
    elif num <= HarryPotter_SampleSpace['Quidditch Game Played'] + HarryPotter_SampleSpace['Bufferfly Collected'] + HarryPotter_SampleSpace['House Cup Won']:
        simulated_results.append('House Cup Won')
    else:
        simulated_results.append('Marauder\'s Map Found')
```

We can see that the probability of obtaining Marauder's Map Found is 0.4, however this may not be the result of the simulation. The reason why the simulation may not give us the same result is because it is based on random selection therefore we may not always get the same result each time we simulate the sample space.
